```python
from collections import Counter
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from sklearn.preprocessing import label_binarize
from sklearn.feature_selection import SelectPercentile
import pandas as pd
import numpy as np
import joblib

# Replace 'QueenslandFLD_target' with your actual target variable
y = QueenslandFLD_target

score_df = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F-Score', 'AUC'])

# Set the number of random subspaces and percentage of features to use
num_subspaces = 2  # Number of random subspaces
feature_percentage = 50  # Percentage of features to use in each subspace

for X, n in zip(df_list, df_names):
    # split into train test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)

    # Apply feature selection to reduce the number of features
    feature_selector = SelectPercentile(percentile=feature_percentage)
    X_train_selected = feature_selector.fit_transform(X_train, y_train)
    X_test_selected = feature_selector.transform(X_test)

    if X_train_selected.shape[1] == 0:
        # Skip model training if no features are selected
        continue

    # L1 regularized logistic regression
    lr_l1 = LogisticRegressionCV(solver='liblinear').fit(X_train_selected, y_train)

    # Random Subspace classifier
    classifiers = []  # List to store the base classifiers
    feature_indices = []  # List to store the feature indices used for each classifier

    for _ in range(num_subspaces):
        features = np.random.choice(range(X_train_selected.shape[1]), size=int(X_train_selected.shape[1] / 2), replace=False)
        X_subset = X_train_selected[:, features]
        clf = LogisticRegression().fit(X_subset, y_train)
        classifiers.append(clf)
        feature_indices.append(features)

    # Predict the class for each classifier and combine the predictions
    y_preds = []
    for clf, features in zip(classifiers, feature_indices):
        X_test_subset = X_test_selected[:, features]
        y_pred = clf.predict(X_test_subset)
        y_preds.append(y_pred)

    # Combine the predictions using majority voting
    y_pred_combined = np.apply_along_axis(lambda x: Counter(x).most_common(1)[0][0], axis=0, arr=np.array(y_preds))

    # Precision, recall, f-score from the multi-class support function
    precision, recall, fscore, _ = score(y_test, y_pred_combined, average='weighted')

    # The usual way to calculate accuracy
    accuracy = accuracy_score(y_test, y_pred_combined)

    # ROC-AUC scores can be calculated by binarizing the data
    auc = roc_auc_score(label_binarize(y_test, classes=[0, 1]),
                        label_binarize(y_pred_combined, classes=[0, 1]),
                        average='weighted')

    # Last, the confusion matrix
    cm = confusion_matrix(y_test, y_pred_combined)

    # Append metrics to the score_df DataFrame
    score_df = score_df.append({'Model': n, 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall,
                                'F-Score': fscore, 'AUC': auc},
                               ignore_index=True)

    # Print the score_df DataFrame
    print(score_df)
```
